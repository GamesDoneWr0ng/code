```python
import matplotlib.pyplot as plt
import numpy as np

def woble(x, y):
    d = np.sqrt(x**2 + y**2)
    return np.sin(d)/d

ax = plt.axes(projection='3d')
lin = np.linspace(-3*np.pi, 3*np.pi, 500)
x, y = np.meshgrid(lin,lin)
z = woble(x,y)

ax.plot_surface(x,y,z, cmap=plt.cm.YlGnBu_r)
```

```python
import matplotlib.pyplot as plt
import numpy as np

wingSize = 0.35
circlesAmount=4
def wing(t, r):
    t = np.deg2rad(t)
    r = np.deg2rad(r)

    circle = wingSize * np.e**np.cos(t)
    radius = (circle - np.cos(circlesAmount * t)) * np.sin(r*circlesAmount)

    x = np.sin(t) * radius * np.cos(r)
    z = np.cos(t) * radius
    y = np.sin(r) * radius

    return x,y,z

t, r = np.meshgrid(np.linspace(0,360,180), np.linspace(0,360,180))
x, y, z = wing(t, r)
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,z)
plt.show()
```
